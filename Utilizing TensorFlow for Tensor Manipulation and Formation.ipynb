# Pemanfaatan TensorFlow dalam Manipulasi dan Pembentukan Tensor


```python
# Mengimpor pustaka TensorFlow dan memberikan alias 'tf'

import tensorflow as tf
```

```python
# Membuat konstanta dengan TensorFlow
con = tf.constant([[1, 2, 3, 4]], shape=[2, 2], dtype=tf.float32)

# Mencetak nilai dari konstanta 'con'
print(con)
```

```python
#Menampilkan Atribut 
print("value of the constant ", con.numpy)
print("type of the constant ", con.dtype)
print("shape of the constant ", con.shape)
print("device of the constant ", con.device)
```

```python
# Membuat tensor yang berisi nol dengan bentuk yang sama seperti 'con'
zeros_like_c = tf.zeros_like(con)

# Mengambil nilai tensor 'zeros_like_c' dalam bentuk NumPy array dan mencetaknya
print(zeros_like_c.numpy())
```

```python
# Membuat tensor 3x3 yang diisi dengan angka 8
fill_d = tf.fill([3, 3], 8)

# Mengambil nilai tensor 'fill_d' dalam bentuk NumPy array dan mencetaknya
print(fill_d.numpy())
```

```python
# Membuat tensor 5x5 dengan angka-angka acak terdistribusi normal
# dengan mean (rerata) 0, stddev (deviasi standar) 1.0, dan seed (benih) 2
random_e = tf.random.normal([5, 5], mean=0, stddev=1.0, seed=2)

# Mengambil nilai tensor 'random_e' dalam bentuk NumPy array dan mencetaknya
print(random_e.numpy())
```

```python
# Membuat list 'list_f'
list_f = [1, 2, 3, 4, 5, 6]

# Menampilkan tipe data dari 'list_f'
print(type(list_f))
```

```python
list_f
```

```python
# Mengonversi list 'list_f' menjadi tensor dengan tipe data 'tf.float32'
tensor_f = tf.convert_to_tensor(list_f, dtype=tf.float32)

# Menampilkan tensor 'tensor_f'
print(tensor_f)
```

```python
# Membuat variabel TensorFlow 'var_1' yang berisi tensor dengan angka 1
# dan bentuk [2, 3]
var_1 = tf.Variable(tf.ones([2, 3]))

# Menampilkan variabel 'var_1'
print(var_1)
```

```python
# Mencetak isi dari variabel TensorFlow 'var_1' dengan menggunakan 'read_value()'
print("Isi dari variabel 'var_1':", var_1.read_value())
```

```python
# Mengubah nilai dari variabel TensorFlow 'var_1' dengan menggunakan 'assign()'
var_value_1 = [[1, 2, 3], [4, 5, 6]]
var_1.assign(var_value_1)

# Mencetak nilai variabel 'var_1' setelah perubahan
print("Nilai 'var_1' setelah diubah:", var_1.read_value())
```

```python
# Menambahkan nilai dari variabel TensorFlow 'var_1' dengan menggunakan 'assign_add()'
var_1.assign_add(tf.ones([2, 3]))

# Mencetak nilai variabel 'var_1' setelah penambahan
print("Nilai 'var_1' setelah penambahan:", var_1)
```

```python
# Membuat tensor 'tensor_h' dengan bentuk [4, 100, 100, 3]
# yang berisi angka-angka acak terdistribusi normal
tensor_h = tf.random.normal([4, 100, 100, 3])

# Menampilkan nilai dari tensor 'tensor_h'
print(tensor_h)
```

```python
# Mengambil potongan dari tensor 'tensor_h' untuk indeks pertama (0) di semua dimensi lainnya
potongan_tensor_h = tensor_h[0, :, :, :]

# Menampilkan nilai dari potongan tensor 'potongan_tensor_h'
print(potongan_tensor_h)
```

```python
# Mengambil setiap elemen kedua (indeks 0 dan 2) pada dimensi pertama (indeks 0)
# dan semua elemen di dimensi lainnya
potongan_tensor_h = tensor_h[::2, ...]

# Menampilkan nilai dari potongan tensor 'potongan_tensor_h'
print(potongan_tensor_h)
```

```python
# Mengambil elemen-elemen tensor 'tensor_h' dalam urutan terbalik pada dimensi pertama (indeks 0)
potongan_tensor_h = tensor_h[::-1]

# Menampilkan nilai dari potongan tensor 'potongan_tensor_h'
print(potongan_tensor_h)
```

```python
# Mengakses elemen dari tensor 'tensor_h' dengan indeks [0, 19, 39, 1]
elemen_tensor_h = tensor_h[0, 19, 39, 1]

# Menampilkan nilai dari elemen tensor yang diakses
print("Nilai elemen tensor:", elemen_tensor_h)
```

```python
# Daftar indeks yang akan digunakan untuk pengumpulan
indices = [0, 1, 3]

# Menggunakan tf.gather untuk mengumpulkan elemen-elemen dari 'tensor_h'
# pada dimensi pertama (indeks 0) dengan indeks yang ada dalam 'indices'
# axis=0 mengindikasikan dimensi yang akan digunakan untuk pengumpulan
# batch_dims=1 mengindikasikan dimensi batch
hasil_gather = tf.gather(tensor_h, axis=0, indices=indices, batch_dims=1)

# Menampilkan hasil pengumpulan
print(hasil_gather)
```

```python
# Membuat konstanta TensorFlow 'reshape_1' dengan bentuk awal [2, 3]
reshape_1 = tf.constant([[1, 2, 3], [4, 5, 6]])

# Menampilkan nilai konstanta 'reshape_1'
print("Nilai 'reshape_1':")
print(reshape_1)

# Mengubah bentuk konstanta 'reshape_1' menjadi [3, 2]
reshaped_tensor = tf.reshape(reshape_1, (3, 2))

# Menampilkan hasil pengubahan bentuk
print("Hasil pengubahan bentuk:")
print(reshaped_tensor)
```

```python
# Membuat tensor 'expand_sample_1' dengan bentuk [100, 100, 3] dan angka-angka acak
expand_sample_1 = tf.random.normal([100, 100, 3], seed=1)

# Menampilkan bentuk (shape) dari tensor 'expand_sample_1'
print("Bentuk (shape) dari 'expand_sample_1':", expand_sample_1.shape)
```

Kesimpulan :

1. Membuat Tensor: Dalam kasus ini, kita dapat membuat tensor dengan berbagai bentuk dan isi menggunakan TensorFlow. Contohnya adalah membuat tensor berisi angka acak terdistribusi normal, angka-angka tertentu, atau menggunakan fungsi tf.constant dan tf.fill.

2. Mengubah Bentuk Tensor: Kita dapat mengubah bentuk tensor menggunakan fungsi tf.reshape. Ini memungkinkan kita untuk mengorganisasi ulang elemen-elemen dalam tensor tanpa mengubah nilainya.

3. Mengakses Elemen Tensor: Elemen-elemen dalam tensor dapat diakses dengan mengindeksnya menggunakan tanda kurung siku. Kita dapat mengambil elemen tunggal atau potongan (slicing) dari tensor sesuai dengan kebutuhan.

4. Variabel TensorFlow: Variabel TensorFlow (dengan tf.Variable) digunakan untuk menyimpan parameter dan bobot dalam model machine learning. Variabel ini dapat diubah nilainya selama pelatihan model.

5. Menggunakan Fungsi tf.gather: tf.gather digunakan untuk mengumpulkan elemen-elemen dari tensor berdasarkan indeks yang ditentukan. Dimensi yang digunakan untuk pengumpulan dapat dikontrol menggunakan axis dan batch_dims.

6. Menggunakan Fungsi tf.squeeze: tf.squeeze digunakan untuk menghapus dimensi dengan ukuran 1 yang mungkin tidak diperlukan dalam tensor.

Selain itu, dalam case study ini kita juga telah memahami bagaimana mencetak nilai dari tensor serta beberapa atribut seperti bentuk (shape), tipe data (dtype), dan sebagainya.
